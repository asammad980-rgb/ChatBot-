```python
import os
import google.generativeai as genai
from google.colab import userdata

API_KEY=userdata.get('YOUR API KEY')
if not API_KEY:
  raise RuntimeError(
      "Gemini API key not found.\n"
      "Set it using:\n"
      "export GEMINI_API_KEY='YOUR_API_KEY'\n"
      "OR in colab:\n"
      "from google colab import userdata\n"
      "userdata.set('GEMINI_API_KEY','YOUR_API_KEY')"
  )
genai.configure(api_key=API_KEY)
model = genai.GenerativeModel("models/gemini-flash-latest")

def chatbot():
  print("\n🤖 Simple Gemini Chatbot")
  print("Type 'exit' to stop the chat.\n")

  while True:
    user_input=input("You: ").strip()
    if user_input.lower()=="exit":
      print("Chatbot stopped.")
      break
    try:
      response=model.generate_content(user_input)
      print("Bot:",response.text.strip())

    except Exception as e:
      print("Error:",e)

if __name__=="__main__":
  chatbot()
```
